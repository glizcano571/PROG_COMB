##Crea un programa que muestre la tabla de multiplicar de un número hasta el 10.

for

```python
numero = int(input("¿Qué tabla quieres ver? "))

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

WHILE

```python
numero = int(input("¿Qué tabla quieres ver? "))
i = 1

while i <= 10:
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
    i += 1 # ¡si esto no esta,bucle infinito!
```

##Desarrolla un programa que calcule el factorial de un número. Recuerda que el factorial de 0 es 1 y no existen factoriales negativos.

for

```python
n = int(input("Introduce un número para calcular su factorial: "))

if n < 0:
    print("Error: No existen factoriales de números negativos.")
elif n == 0:
    print("El factorial de 0 es 1.")
else:
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    print(f"El factorial de {n} es {factorial}")
```

while

```python
n = int(input("Introduce un número para calcular su factorial: "))

if n < 0:
    print("Error: No existen factoriales de números negativos.")
elif n == 0:
    print("El factorial de 0 es 1.")
else:
    factorial = 1
    i = 1
    # Mientras i sea menor o igual a n
    while i <= n:
        factorial *= i # Acumular el resultado
        i += 1         # Súbimos el contador
    print(f"El factorial de {n} es {factorial}")
```

##Crea un programa que muestre los primeros n números de la serie Fibonacci.



for

```python
n = int(input("¿Cuántos números de Fibonacci quieres ver? "))

a, b = 0, 1
for _ in range(n):
    print(a, end=" ")
    a, b = b, a + b
```

while

```python
n = int(input("¿Cuántos números de Fibonacci quieres ver? "))

a, b = 0, 1
contador = 0

while contador < n:
    print(a, end=" ")
    a, b = b, a + b
    contador += 1
```

##20. Escribe un programa que adivine entre un conjunto de diez colores definidos por el usuario, con un máximo de 10 intentos.

for

```python
colores=[]
for i in range(10):
  color=input("Escribe un color: ")
  colores.append(color)
secreto = colores[0]
for i in range(10):
  intento=input("Adivina el color: ")
  if intento==secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
print("El color era:", secreto)
```

while

```python
colores=[]
i=0
while i < 10:
  color = input("Escribe un color: ")
  colores.append(color)
  i=i+1
secreto = colores[0]
intentos = 0
while intentos < 10:
  intento = input("Adivina el color: ")
  if intento == secreto:
      print("¡Ganaste!")
      break
  else:
      print("No es ese color")
  intentos = intentos + 1
print("El color era:", secreto)
```

##22. Crea un programa que calcule la suma de los dígitos de un número.

for

```python
suma=0
num=input("Ingresa un número: ")
for digito in num:
    suma+=int(digito)
print("La suma de los dígitos es:", suma)
```

while

```python
suma=0
num=int(input("Ingresa un número: "))
while num>0:
  digito=num % 10
  suma+=digito
  num=num//10
print("La suma de los dígitos es:", suma)
```
