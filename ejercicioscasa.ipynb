## Crea un programa que determine si un triángulo es equilátero, isósceles o escaleno según sus lados.

## Desarrolla un programa que simule una calculadora simple (+, -, *, /) con dos números.

```python
n1 = float(input("Primer número: "))
n2 = float(input("Segundo número: "))
operacion = input("Elige operación (+, -, *, /): ")
```

```python
if operacion == "+":
    print("Resultado:", n1 + n2)
elif operacion == "-":
    print("Resultado:", n1 - n2)
elif operacion == "*":
    print("Resultado:", n1 * n2)
elif operacion == "/":
    if n2 != 0:
        print("Resultado:", n1 / n2)
    else:
        print("Error")
else:
    print("Operación no válida.")
```

## Desarrolla un programa que valide si tres números pueden formar un triángulo y de qué tipo es.

```python
a = float(input("Lado A: "))
b = float(input("Lado B: "))
c = float(input("Lado C: "))
```

```python
if (a + b > c) and (a + c > b) and (b + c > a):
    # Si entra aquí, determinamos el tipo
    if a == b == c:
        print("Equilátero")
    elif a == b or a == c or b == c:
        print("Isósceles")
    else:
        print("Escaleno")
else:
    print("los lados no pueden formar un triángulo")
```
