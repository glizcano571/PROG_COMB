```python
'''
Andres lizcano
nikold batuista
camilo zambrano
'''
print("polimorfismo simple")

class Perro:
    def hablar(self):
        return "Guau! Guau!"

class Gato:
    def hablar(self):
        return "Miau! Miau!"

class Vaca:
    def hablar(self):
        return "Muuuuu!"

print("DEMOSTRACIÓN DE POLIMORFISMO")

granja = [Perro(), Gato(), Vaca(), Perro()]
for animal in granja:
    print("El animal dice:", animal.hablar())
    print("No nos importa qué animal es, solo que sabe .hablar()")

print("el polimorifsmo simple es la capacidad de objetos de distintas clases de responder al mismo mensaje o método de formas diferentes.")
```
