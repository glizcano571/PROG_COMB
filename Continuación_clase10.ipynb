#Programación orientada a objetos POO
-Paradigma de programación enfocado en crear marcos (clase) que a su vez generan instancias (objetos).
- Está fundamentada bajo 4 principios fundamentales o pilares: **Herencia, Abstracción, Polimorfismo y Encapsulamiento**
- Se enfoca en crear plantillas (class) las cuales desarrollan instancias (objetos). Esos objetos hacen cosas (métodos), y tienen características propias (atributos).
- Ventajas al implementar POO
1. Escalabilidad: un mismo código en diferentes app.
2. Seguridad: No cualquier persona tiene acceso al código.
3. Modularidad: El código se expresa a partir de "pedazos" (métodos).
4. Reusabilidad: Se puede reutilizar partes del código en lugar de generar un código nuevo.
5. Mantenibilidad: El código puede ser fácilmente corregido y actualizado.

Creando clases

````python
class Edificio:
  #Constructor: Para crear atributos, __init__/(parámetros)
  def __init__(self,nombre,habitación,dirección):
    self.nombre=nombre
    self.habitación=habitación
    self.dirección=dirección
  #Métodos son las mismas funciones
  def mostrar_info(self):
    print(f"Nombre:{self.nombre}")
    print(f"Dirección: {self.dirección}")
    print(f"No. de habitaciones: {self.habitación}")

class Floristería(Edificio):
  def __init__(self, nombre, habitación, dirección, tipo_flores):
    super().__init__(nombre, habitación, dirección)
    self.tipo_flores=tipo_flores

  def vender_flor(self, flor):
    if flor in self.tipo_flores:
      print(f"Se vendió una flor de tipo {flor}")
    else:
      print(f"No tengo de esa flor")
````

Creando objetos (instancia de clase)


```
variable=clase(parámetros)
```



````python
mi_floristería=Floristería(
  "Floristería El chamo",
  4,
  "Carrera 33, 28-15 - Bucaramanga",
  ["rosa", "tulipán", "lirio"]
)
````

````python
mi_floristería.mostrar_info()
````
