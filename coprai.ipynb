```python
# EJERCICIO 1
frutas = {"nombre": ["manzana", "pera", "uva"],
          "precio": [2500, 1800, 3200]}
print(frutas["precio"][0])

# EJERCICIO 2
personajes = {"personajes": ["nikold", "andres", "camilo"],
              "nivel": [4, 2, 3]}
personajes["personajes"].append("nuevo")
personajes["nivel"].append(1)
print(personajes)

# EJERCICIO 3
pociones = {"nombre": ["salud", "mana", "fuerza"],
            "cantidad": [10, 5, 8]}
pociones["cantidad"][0] = 15
print(pociones)

# EJERCICIO 5
estudiantes = {"nombre": ["Juan", "Maria", "Pedro"],
               "nota": [8.5, 9.0, 7.5]}

def promedio(estudiantes):
    total = sum(estudiantes["nota"])
    return total / len(estudiantes["nota"])

def get_estudiante(estudiantes, nombre):
    indice = estudiantes["nombre"].index(nombre)
    return estudiantes["nota"][indice]

print(promedio(estudiantes))
print(get_estudiante(estudiantes, "Maria"))

# EJERCICIO 6
personajes = {"nombre": ["hechicero", "guerrero", "mago"],
              "poder": [10, 12, 15],
              "nivel": [8, 6, 9]}

if "hechicero" in personajes["nombre"]:
    indice = personajes["nombre"].index("hechicero")
    print(f"Poder: {personajes['poder'][indice]}, Nivel: {personajes['nivel'][indice]}")
else:
    print("No existe")

# EJERCICIO 7
palabra = "magia"
conteo = {"letra": [],
          "veces": []}
for letra in palabra:
    if letra in conteo["letra"]:
        indice = conteo["letra"].index(letra)
        conteo["veces"][indice] += 1
    else:
        conteo["letra"].append(letra)
        conteo["veces"].append(1)
print(conteo)

# EJERCICIO 8
monstruos = ["slime", "dragon", "slime", "lobo", "dragon", "slime"]
conteo = {"monstruo": [],
          "cantidad": []}
for monstruo in monstruos:
    if monstruo in conteo["monstruo"]:
        indice = conteo["monstruo"].index(monstruo)
        conteo["cantidad"][indice] += 1
    else:
        conteo["monstruo"].append(monstruo)
        conteo["cantidad"].append(1)
print("Conteo final:", conteo)

# EJERCICIO 9
flores = {"nombre": ["rosa", "tulipan", "girasol"],
          "cantidad": [50, 30, 40],
          "precio": [2000, 1500, 1800]}
print(flores)

# EJERCICIO 10
inventario = {"distrito": ["FUNPRO", "PROGCOM", "SEMING", "MOBA"],
              "objeto": [["pociones", "armas"], ["libros", "cristales"], ["semillas", "fertilizante"], ["monedas", "joyas"]],
              "cantidad": [[20, 10], [50, 15], [100, 25], [5000, 8]]}

print(inventario)
```
