```python
datos=pd.read_csv('saludmental.csv')
```

```python
datos.head()
```

```python
# Agrupamos por nombre de diagnóstico y sumamos la columna 'Total'
top_5_diagnosticos = datos.groupby('diagnostico_ingreso')['Total'].sum().nlargest()
#TOTAL DE CASOS POR AÑO
# Agrupamos por la columna 'Año diagnóstico' y sumamos
casos_por_anio = datos.groupby('Año diagnóstico')['Total'].sum()

# Mostrar Resultados
print("Los 5 diagnósticos más frecuentes son:")
print(top_5_diagnosticos)

print("\nTotal de casos por año:")
print(casos_por_anio)
```
