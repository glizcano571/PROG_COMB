##Crea un programa que convierta grados Celsius a Fahrenheit usando la fórmula: °F = (°C × 9/5) + 32

```python
# 1. Pedimos datos al usuario
celsius = float(input("Introduce los grados Celsius: "))  #(convertimos a float para permitir decimales)
fahrenheit = (celsius * 9 / 5) + 32
print(f"{celsius}°C equivalen a {fahrenheit}°F")
```

##Desarrolla un programa que calcule el promedio de tres números ingresados por el usuario.

```python
num1 = float(input("Introduce tu primer número: "))
num2 = float(input("Introduce tu segundo número: "))
num3 = float(input("Introduce tu tercer número: "))

promedio = (num1 + num2 + num3) / 3
print("EL PROMEDIO ES ", promedio)
```

##Escribe un programa que calcule el perímetro y área de un círculo dado su radio.

```python
radio = float(input("Introduce el radio: "))

# 2. Definimos pi de forma sencilla
pi = 3.1416

# 3. Calculamos
perimetro = 2 * pi * radio
area = pi * (radio * radio)

# 4. Mostramos resultados
print("Perímetro:", perimetro)
print("Área:", area)
```
